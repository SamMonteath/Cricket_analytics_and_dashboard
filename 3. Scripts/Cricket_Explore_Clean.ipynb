```python
#Making that basic view copy and paste able
#First define the data type in each columns

column_types = pd.DataFrame({
    'Column Name': df.columns,
    'Data Type': df.dtypes.astype(str).values
})
```

```python
print(column_types)
```

```python
# And then adding the count of non-null values

summary = pd.DataFrame({
    'Column Name': df.columns,
    'Data Type': df.dtypes.astype(str).values,
    'Non-Null Count': df.notnull().sum().values
})
```

```python
print(summary)
```

```python
#Producing a .csv

summary.to_csv('column_summary.csv', index=False)
```
